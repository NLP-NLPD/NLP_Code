```python
import re
import PyPDF2

# PDF 파일 읽기
file_path = '/Users/whovi/OneDrive/Documents/ewha.pdf'
with open(file_path, 'rb') as file:
    pdf_reader = PyPDF2.PdfReader(file)
    pdf_text = ""
    for page in pdf_reader.pages:
        page_text = page.extract_text()
        if page_text:
            pdf_text += page_text

# "제n장 제목"을 기준으로 텍스트 분할
chapter_pattern = r"(제\d+장\s*(?:.(?!제\d+장))*)(.*?)(?=제\d+장|$)"
chapters = re.findall(chapter_pattern, pdf_text, re.DOTALL)

# 데이터베이스 형식으로 각 장 저장
document_sections = {}
for chapter_title, chapter_content in chapters:
    chapter_title = chapter_title.strip()  
    chapter_content = chapter_content.strip()
    
    # 각 장 내에서 "제n조"를 기준으로 텍스트 분할
    article_pattern = r"(제\d+조(?:의\d+)?)"
    articles = re.split(article_pattern, chapter_content)
    
    # 각 장의 섹션 저장
    articles_dict = {}
    for j in range(1, len(articles), 2):
        article_title = articles[j].strip()  
        article_content = articles[j+1].strip()
        articles_dict[article_title] = article_content
    
    document_sections[chapter_title] = articles_dict

# 결과 출력 (조의 내용 포함)
for chapter, articles in document_sections.items():
    print(f"Section: {chapter}")
    for article, content in articles.items():
        print(f"  {article}")
        print(f"    내용: {content}\n")
    print("\n")  
```
